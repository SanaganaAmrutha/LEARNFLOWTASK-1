```python
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
import matplotlib.pyplot as plt

# Define the directory paths
train_dir = '/home/amrtutha/Downloads/ML task 1(1)/images'
test_dir = '/home/amrtutha/Downloads/ML task 1(1)/images'

# Ensure the directory exists
if not os.path.exists(train_dir):
    raise FileNotFoundError(f"Directory '{train_dir}' does not exist. Please check the path and try again.")

# Check for the presence of subdirectories and images
subdirs = [d for d in os.listdir(train_dir) if os.path.isdir(os.path.join(train_dir, d))]
if not subdirs:
    raise ValueError(f"No subdirectories found in '{train_dir}'. Ensure the directory structure is correct.")
for subdir in subdirs:
    subdir_path = os.path.join(train_dir, subdir)
    if not os.listdir(subdir_path):
        raise ValueError(f"No images found in subdirectory '{subdir_path}'.")

# Define the data generators
train_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

validation_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)

# Define the model
num_classes = len(subdirs)

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(128, activation='relu'),
    Dense(num_classes, activation='softmax')  # Adjust to the number of classes
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# Display one of the training images
images, labels = next(train_generator)
plt.imshow(images[0])
plt.title(f'Class: {labels[0]}')
plt.show()

# Train the model
history = model.fit(train_generator, epochs=20, validation_data=validation_generator)

# Evaluate the model on the test data
if os.path.exists(test_dir):
    test_generator = train_datagen.flow_from_directory(
        test_dir,
        target_size=(128, 128),
        batch_size=32,
        class_mode='categorical'
    )
    loss, accuracy = model.evaluate(test_generator)
    print(f'Test accuracy: {accuracy}')
else:
    print(f"Test directory '{test_dir}' does not exist.")
```
